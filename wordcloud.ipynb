```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def remove_punctuation(text):
    text=str(text)
    import string
    punct = string.punctuation + '.،!"#$%&|×/:؛,][\}{«»<>؟'

    return text.translate({ord(p): " " for p in punct})
```

```python
import re
def remove_enter (comment):
    comment =re.sub(r'[(\r\n)+]|[\r\n]+', ' ', comment)
    comment =re.sub(r'\s+', ' ', comment)
    return comment
```

```python
import persian_wordcloud
from persian_wordcloud.wordcloud import PersianWordCloud, add_stop_words
```

```python
# Add another stopword
stopwords = add_stop_words(['است'])
```

```python
def removeWeirdChars(text):
    weridPatterns = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u'\U00010000-\U0010ffff'
                               u"\u200d"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\u3030"
                               u"\ufe0f"
                               u"\u2069"
                               u"\u2066"
                               u"\u200c"
                               u"\u2068"
                               u"\u2067"
                               "]+", flags=re.UNICODE)
    return weridPatterns.sub(r'', text)
```

```python
hamText=removeWeirdChars(hamText)
text = hamText
```

```python
# Generate a word cloud image
wordcloud = PersianWordCloud(
    only_persian=True,
    max_words=100,
    stopwords=stopwords,
    margin=0,
    width=800,
    height=800,
    min_font_size=1,
    max_font_size=500,
    background_color="white"
).generate(text)
```

```python
image = wordcloud.to_image()
image.show()
```

```python
image.save('wordcloud.png')
```
